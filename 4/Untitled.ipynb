```python
y = [3244,3899,4559,2573,1428,530,101,50]
x = range(8)
plt.bar(x,y)
```

```python
s = sum(y)
equalized_x= []
fraction = 7./s
ss = 0
for i in y:
    ss += i
    v = ss*fraction
    equalized_x.append(round(v))
```

```python
equalized_x = map(int,equalized_x)
equalized_x
```

```python
new_y =np.zeros(8)
for i,v in enumerate(equalized_x):
    new_y[v] += y[i]
plt.bar(x,new_y)
    
```
