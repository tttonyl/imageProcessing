```python
import matplotlib.pyplot as plt
import skimage.io as io
import numpy as np
import math
% matplotlib inline
```

```python
T = np.array(range(256),dtype = np.uint8)*2
```

```python
c = io.imread("cameraman.tif")
c_complements = 256 - c
plt.imshow(c_complements,vmin = 0, vmax =255, cmap= plt.cm.gray)
```

```python
hist_value = []
for i in range(256):
    counts = np.count_nonzero(c == i)
    hist_value.append(counts)
```

```python
plt.bar(range(256),hist_value)
```

```python
y = [3244,3899,4559,2573,1428,530,0,50]
x = range(8)
plt.bar(x,y)
```

```python
s = sum(y)
new_y= []
fraction = 7./s
ss = 0
for i in y:
    ss += i
    v = ss*fraction
    new_y.append(round(v))
```

```python
plt.bar(x,new_y)
```
